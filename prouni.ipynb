```python
formato_padrao = "%Y-%m-%d"
prouni['Nascimento'] = pd.to_datetime(prouni['Nascimento'], format=formato_padrao)
```

```python

def verificar_formato(data):
    try:
        pd.to_datetime(data)
        return True
    except ValueError:
        return False

formato_correto = prouni['Nascimento'].apply(verificar_formato).all()

if formato_correto:
    print("Todas as datas estão no mesmo formato.")
else:
    print("As datas não estão todas no mesmo formato.")
```
