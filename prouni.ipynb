```python
cursos_sociais_humanas = ['Sociologia', 'Antropologia', 'Politica', 'Historia', 'Geografia', 'Linguistica', 'Filosofia', 'Arqueologia', 'Relacoes Internacionais', 'Direito']
cursos_naturais_exatas = ['Fisica', 'Quimica', 'Biologia', 'Matematica', 'Astronomia', 'Geologia']
artes_humanidade = ['Artes Visuais', 'Teatro', 'Musica', 'Danca', 'Literatura', 'Cinema', 'Estudos Culturais', 'Estudos de Midia', 'Design', 'Arquitetura', 'Historia da Arte']
saude = ['Medicina', 'Enfermagem', 'Odontologia', 'Fisioterapia', 'Farmacia', 'Biomedicina', 'Nutricao', 'Psicologia', 'Clinica', 'Saude Publica', 'Ciencias do Esporte']
agrarias = ['Agricultura', 'Agronomia', 'Engenharia Agricola', 'Zootecnia', 'Veterinaria', 'Tecnologia de Alimentos']
engenharias = ['Engenharia Civil', 'Engenharia Mecanica', 'Engenharia Eletrica', 'Engenharia de Computacao', 'Engenharia Quimica', 'Engenharia de Materiais', 'Engenharia de Software', 'Engenharia Ambiental', 'Ciencia da Computacao', 'Tecnologia da Informacao']
sociais_aplicadas = ['Administracao', 'Contabilidade', 'Economia', 'Marketing', 'Comercio Exterior', 'Turismo', 'Comunicacao Social', 'Recursos Humanos', 'Gestao Publica']
educação = ['Pedagogia', 'Licenciatura', 'Educacao Fisica', 'Letras', 'Psicopedagogia', 'Educacao Especial']
ambientais = ['Ecologia', 'Ambiental', 'Ciencias do Ambiente', 'Sustentabilidade']
comunicação = ['Comunicacao Social', 'Jornalismo', 'Publicidade e Propaganda', 'Relacoes Publicas', 'Producao Audiovisual', 'Estudos de Midia']
def categorize_course(course):
    if course in cursos_sociais_humanas:
        return 'Ciências Sociais e Humanas'
    elif course in cursos_naturais_exatas:
        return 'Ciências Naturais e Exatas'
    elif course in artes_humanidade:
        return 'Artes e Humanidades'
    elif course in saude:
        return 'Saúde'
    elif course in agrarias:
        return 'Ciências Agrárias'
    elif course in engenharias:
        return 'Engenharia e Tecnologia'
    elif course in sociais_aplicadas:
        return 'Ciências Sociais Aplicadas'
    elif course in educação:
        return 'Educação'
    elif course in ambientais:
        return 'Ciências Ambientais'
    elif course in comunicação:
        return 'Comunicação'
    else:
        return 'Outros'
```

```python
prouni['AREA'] = prouni['CURSO'].apply(categorize_course)
```

```python
prouni
```

```python
def contar_categorias(df, coluna):
    # Criar um dicionário para armazenar as contagens de cada categoria
    contagem_categorias = {}

    # Iterar sobre cada elemento da coluna
    for categoria in df[coluna]:
        if categoria in contagem_categorias:
            # Incrementar a contagem se a categoria já existe no dicionário
            contagem_categorias[categoria] += 1
        else:
            # Inicializar a contagem para a nova categoria
            contagem_categorias[categoria] = 1

    return contagem_categorias
```

```python
contagem = contar_categorias(prouni, 'AREA')

for categoria, contagem_categoria in contagem.items():
    print(f"Categoria: {categoria} | Quantidade: {contagem_categoria}")
```
